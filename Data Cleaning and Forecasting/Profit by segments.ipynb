```python
#import libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
#read file
profitby_p=pd.read_csv('Coffee_company.csv')
profitby_p['Segment'].head
```

```python
#clean columns
profitby_p.columns=profitby_p.columns.str.replace(' ', '')
```

```python
profitby_p['Profit'].head()
```

```python
#clean the profit data 
profitby_p['Profit']=profitby_p['Profit'].str.replace('$', '').str.replace(',','')
```

```python
#transform data to numeric
profitby_p['Profit']=pd.to_numeric(profitby_p['Profit'], errors='coerce')
```

```python
#group by segment and sum the profit
grp=profitby_p.groupby('Segment')['Profit'].sum().reset_index()
#check grp data
grp.head()
```

```python
#plot
plt.figure(figsize=(10, 5))
plt.bar(x=grp['Segment'], height=grp['Profit'], width=0.1, color='skyblue')
plt.ylabel('Profit in 10s of $ Millions')
plt.title('Segments specific Profit for the period 2021-2023')
for i, value in enumerate(grp['Profit']):
    plt.text(i, value+0.2, f'{(value/10000000):.2f}', ha='center', va='bottom')
plt.show
```
